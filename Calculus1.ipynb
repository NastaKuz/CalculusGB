4.  $$\forall x \in \mathbb{C} \nexists y \in \mathbb{C}, x > y || x < y$$
–î–ª—è –ª—é–±–æ–≥–æ –∫–æ–º–ø–ª–µ–∫—Å–Ω–æ–≥–æ —á–∏—Å–ª–∞ x –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–∫–æ–≥–æ –∫–æ–º–ø–ª–µ–∫—Å–Ω–æ–≥–æ —á–∏—Å–ª–∞ y, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º x –±–æ–ª—å—à–µ y –∏–ª–∏ x –º–µ–Ω—å—à–µ y.

–û—Ç—Ä–∏—Ü–∞–Ω–∏–µ: $$\exists x \in \mathbb{C} \exists y \in \mathbb{C}, x \leq y, x \geq y$$

5.  $$\forall y \in [0; \pi/2] \exists \epsilon > 0: \sin y < \sin (y + \epsilon)$$

–û—Ç—Ä–∏—Ü–∞–Ω–∏–µ: $$\exists y \in [0; \pi/2] \forall \epsilon > 0: \sin y \geq \sin (y + \epsilon)$$
–°—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–∫–æ–π y –æ—Ç 0 –¥–æ pi/2 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –ø—Ä–∏ –ª—é–±–æ–º epsilon –±–æ–ª—å—à–µ 0 –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è sin(y) –±–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ sin(y + epsilon).

6. $$\forall y \in [0; \pi) \exists \epsilon > 0: \cos y > \cos (y + \epsilon)$$
–î–ª—è –ª—é–±–æ–≥–æ y –æ—Ç 0 –¥–æ pi (–Ω–µ –≤–∫–ª—é—á–∞—è pi) —Å—É—â–µ—Å—Ç–≤—É–µ—Ç epsilon –±–æ–ª—å—à–µ 0, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º cos(y) –±–æ–ª—å—à–µ cos(y + epsilon).  
–í—ã—Ä–∞–∂–µ–Ω–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ, —Ç–∞–∫ –∫–∞–∫ –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –Ω–µ –≤–∫–ª—é—á–µ–Ω pi, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º cos –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. –î–ª—è –≤—Å–µ—Ö –¥—Ä—É–≥–∏—Ö —Å–ª—É—á–∞–µ–≤ –ø—Ä–∏ –¥–≤–∏–∂–µ–Ω–∏–∏ –ø–æ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–º—É –∫—Ä—É–≥—É –∫–æ—Å–∏–Ω—É—Å —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è.  
–û—Ç—Ä–∏—Ü–∞–Ω–∏–µ:  $$\exists y \in [0; \pi) \forall \epsilon > 0: \cos y \leq \cos (y + \epsilon)$$
–°—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–∫–æ–π y –æ—Ç 0 –¥–æ pi (–Ω–µ –≤–∫–ª—é—á–∞—è pi), –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –ø—Ä–∏ –ª—é–±–æ–º epsilon –±–æ–ª—å—à–µ 0 –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è cos(y) –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ cos(y + epsilon).  
–í—ã—Ä–∞–∂–µ–Ω–∏–µ –ª–æ–∂–Ω–æ, –ø—Ä–∏ y=0 cosy = 1, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∫–æ—Å–∏–Ω—É—Å–∞.


7. $$\exists x: x \notin \{ \mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}, \mathbb{C} \} $$
–°—É—â–µ—Å—Ç–≤—É–µ—Ç x, –Ω–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏–π –º–Ω–æ–∂–µ—Å—Ç–≤–∞–º –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö, —Ü–µ–ª—ã—Ö, —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö, –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏–ª–∏ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–µ–ª.  
–í—ã—Ä–∞–∂–µ–Ω–∏–µ –ª–æ–∂–Ω–æ, —Ç–∞–∫ –∫–∞–∫ –ª—é–±–æ–µ —á–∏—Å–ª–æ x –¥–æ–ª–∂–Ω–æ –≤—Ö–æ–¥–∏—Ç—å —Ö–æ—Ç—è –±—ã –≤ –æ–¥–Ω—É –∏–∑ —ç—Ç–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π.  
–û—Ç—Ä–∏—Ü–∞–Ω–∏–µ: $$\forall x: x \in \{ \mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}, \mathbb{C} \} $$
–õ—é–±–æ–π x –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–∞–º –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö, —Ü–µ–ª—ã—Ö, —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö, –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏–ª–∏ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–µ–ª.  
–í—ã—Ä–∞–∂–µ–Ω–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ, –ª—é–±–æ–µ —á–∏—Å–ª–æ x –¥–æ–ª–∂–Ω–æ –≤—Ö–æ–¥–∏—Ç—å —Ö–æ—Ç—è –±—ã –≤ –æ–¥–Ω—É –∏–∑ —ç—Ç–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π.  

–î–∞–Ω—ã —Ç—Ä–∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ a, b –∏ —Å. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤—Å–µ –∏–∑—É—á–µ–Ω–Ω—ã–µ –≤–∏–¥—ã –±–∏–Ω–∞—Ä–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ü–∏–π –Ω–∞–¥ –≤—Å–µ–º–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è–º–∏ –º–Ω–æ–∂–µ—Å—Ç–≤.

*–í—ã–ø–æ–ª–Ω–∏—Ç—å –∑–∞–¥–∞–Ω–∏–µ 2 –Ω–∞ —è–∑—ã–∫–µ Python

```python
import random 

random.seed(42)
a = set([random.randint(0, 99) for _ in range(random.randint(1, 10))])
b = set([random.randint(0, 99) for _ in range(random.randint(1, 10))])
c = set([random.randint(0, 99) for _ in range(random.randint(1, 10))])

sets = {"a": a,
        "b": b,
        "c": c}
```

```python
import itertools

for combination in itertools.combinations(sets.keys(), 2):
  x, y = combination[0], combination[1]
  print(f'–ò—Å—Ö–æ–¥–Ω—ã–µ –º–Ω–æ–∂–µ—Å—Ç–≤–∞: {x}, {y}')
  print(f'x: {sets[x]},\ny: {sets[y]}')
  print(f'–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ {x} | {y}: {sets[x] | sets[y]}')
  print(f'–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ {x} & {y}: {sets[x] & sets[y]}')
  print(f'–†–∞–∑–Ω–æ—Å—Ç—å {x} - {y}: {sets[x] - sets[y]}')
  print(f'–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å {x} ^ {y}: {sets[x] ^ sets[y]}')
  
  cartesian = [(a,b) for a in sets[x] for b in sets[y]]
  print(f'–î–µ–∫–∞—Ä—Ç–æ–≤–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ {x} * {y}: {set(cartesian)}')

  print('_____________')
```
