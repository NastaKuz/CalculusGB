4. Найти предел:
 $$\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}= (\frac{0}{0})=\lim_{x \to 0} \frac{x^2(3x - 2)}{4x^2} = \lim_{x \to 0} \frac{3x - 2}{4} = -\frac{2}{4} = -\frac{1}{2}$$
