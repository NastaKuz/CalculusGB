# Тема “Предел функции”

Найти предел:
$$\lim_{x \to \infty} {(\frac{x+3}{x})}^{4x+1} = \lim_{x \to \infty} {(1+\frac{3}{x})}^{4x+1} =  [1^\infty] = \lim_{x \to \infty} {(1+\frac{3}{x})}^{(\frac{x}{3})*\frac{3}{x}* (4x+1)} = \lim_{x \to \infty}{e^{\frac{3(4x+1)}{x}}} = \lim_{x \to \infty}{e^{12+\frac{1}{x}}} = e^{12}$$

Найти предел:
$$\lim_{x \to 0} \frac{\sin{2x}}{4x} = [\frac{0}{0}] = \lim_{x \to 0} \frac{\sin{2x}}{2x}*\frac{2x}{4x} = \lim_{x \to 0} \frac{\sin{2x}}{2x}*\frac{1}{2} = \frac{1}{2}$$

$$\lim_{x \to 0} \frac{x}{\sin{x}} = [\frac{0}{0}] = \lim_{x \to 0} \frac{1}{\frac{\sin{x}}{x}} = 1$$

$$\lim_{x \to 0} \frac{x}{\arcsin{x}} = [\frac{0}{0}] = \lim_{x \to 0} \frac{x}{x} = 1 $$
(Замена арксинуса на x по таблице эквивалентности)

$$\lim_{x \to \infty} {(\frac{4x+3}{4x-3})}^{6x} = \lim_{x \to \infty} {(\frac{4x-3+3+3}{4x-3})}^{6x} = \lim_{x \to \infty} {(1+\frac{6}{4x-3})}^{6x}=[1^\infty] = \lim_{x \to \infty} {(1+\frac{6}{4x-3})}^{\frac{4x-3}{6}\frac{6}{4x-3}6x} = \lim_{x \to \infty}{e^{\frac{6}{4x-3}6x}} =\lim_{x \to \infty}{e^{\frac{36x}{4x-3}}}= e^{9}$$

$$\lim_{x \to \infty} \frac{\sin{x} +\ln{x}}{x} = \lim_{x \to \infty} \frac{\sin{x}}{x} + \lim_{x \to \infty} \frac{\ln{x}}{x} = 0 + 0 = 0$$

(По скорости роста функций, x растет быстрее чем lnx)

$$\lim_{x \to 0} \frac{\sin{x} +\ln{x}}{x} =\lim_{x \to 0} \frac{\sin{x}}{x} + \lim_{x \to 0} \frac{\ln{x}}{x} = [\ln{x} => \textrm{x положительный, рассматриваем правосторонний предел]} = 1 - \infty = -\infty$$

(Пределы определены по графику)

$(\ln{({2x+1})^3})' = \frac{1}{(2x+1)^3} * 3(2x+2)^2 * 2 = \frac{6}{2x+1}$
